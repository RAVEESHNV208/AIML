### MODELS TRAINING

````python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import lightgbm as lgb
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
````

#### setting up data

````python
df['age_years'] = (df['age'] / 365).astype(int)

# age categories
df['age_category'] = pd.cut(df['age_years'], 
                            bins=[0, 30, 50, 60, np.inf], 
                            labels=['Under 30', '30-49', '50-59', '60 and above'])

# Encode cholesterol and glucose levels
df['cholesterol_category'] = df['cholesterol'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})
df['glucose_category'] = df['gluc'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})

# Calculate and categorize BMI
df['bmi'] = df['weight'] / ((df['height'] / 100) ** 2)
df['bmi_category'] = pd.cut(df['bmi'], 
                            bins=[0, 18.5, 25, 30, np.inf], 
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Categorize blood pressure
def bp_category(systolic, diastolic):
    if systolic < 120 and diastolic < 80:
        return 'Normal'
    elif 120 <= systolic < 130 and diastolic < 80:
        return 'Elevated'
    elif 130 <= systolic < 140 or 80 <= diastolic < 90:
        return 'Hypertension Stage 1'
    elif systolic >= 140 or diastolic >= 90:
        return 'Hypertension Stage 2'
    else:
        return 'Hypertensive Crisis'

df['bp_category'] = df.apply(lambda x: bp_category(x['ap_hi'], x['ap_lo']), axis=1)

df['gender'] = df['gender'].map({1: 'Female', 2: 'Male'})
df['smoke'] = df['smoke'].map({0: 'Non-smoker', 1: 'Smoker'})
df['alco'] = df['alco'].map({0: 'Non-drinker', 1: 'Drinker'})
df['active'] = df['active'].map({0: 'Inactive', 1: 'Active'})
````

````python
columns_to_drop = ['age']
df = df.drop(columns=columns_to_drop)

categorical_columns = ['gender', 'age_category', 'cholesterol_category', 'glucose_category', 
                       'bmi_category', 'bp_category', 'smoke', 'alco', 'active']

# Create a dictionary to store mappings
label_mappings = {}

# Apply LabelEncoder and store mappings for each column
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # Encode the column
    label_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))  # Store mapping

# Print the mappings
for col, mapping in label_mappings.items():
    print(f"Mapping for {col}: {mapping}")
````

````python
x = df.drop(columns=['cardio'])
y = df['cardio']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
````

#### Logistic Regression

````python
# Initialize and train the logistic regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = log_reg.predict(X_test_scaled)
````

````python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# Classification report
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
````

### MAIN MODEL

#### Now training the META Model

````python
#reimporting to make sure nothing left out for further process
import numpy as np
from sklearn.model_selection import cross_val_predict

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

import joblib
````

````python

base_models = [
    ('lgbm', LGBMClassifier(objective='binary', metric='binary_logloss', num_leaves=31, 
                            learning_rate=0.05, feature_fraction=0.9, random_state=42)),
    ('xgb', XGBClassifier(max_depth=6, eta=0.3, objective='binary:logistic', eval_metric='logloss', random_state=42)),
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('lr', LogisticRegression(random_state=42)),
    ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
]

lgbm_params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
}

# Dictionary to store trained models
trained_models = {}

# Generate meta-features
meta_features = np.zeros((x_train.shape[0], len(base_models)))
for i, (name, model) in enumerate(base_models):
    if name == 'lgbm':
        train_data = lgb.Dataset(x_train, label=y_train)
        val_data = lgb.Dataset(x_test, label=y_test)
        lgb_model = lgb.train(lgbm_params, train_data, num_boost_round=100, valid_sets=[val_data])
        meta_features[:, i] = lgb_model.predict(x_train)

    elif name == 'xgb':
        dtrain = xgb.DMatrix(x_train, label=y_train)
        dval = xgb.DMatrix(x_test, label=y_test)
        xgb_params = {
            'max_depth': 6,
            'eta': 0.3,
            'objective': 'binary:logistic',
            'eval_metric': 'logloss'
        }
        xgb_model = xgb.train(xgb_params, dtrain, num_boost_round=100, evals=[(dval, "Val")], early_stopping_rounds=10)
        meta_features[:, i] = xgb_model.predict(xgb.DMatrix(x_train))

    else:
        meta_features[:, i] = cross_val_predict(model, X_train_scaled, y_train, cv=5, method='predict_proba')[:, 1]

# Train base models on full training data
for name, model in base_models:
    if name == 'lgbm':
        train_data = lgb.Dataset(x_train, label=y_train)
        model = lgb.train(lgbm_params, train_data, num_boost_round=100)
        trained_models[name] = model 
    elif name == 'xgb':
        dtrain = xgb.DMatrix(x_train, label=y_train)
        xgb_params = {
            'max_depth': 6,
            'eta': 0.3,
            'objective': 'binary:logistic',
            'eval_metric': 'logloss'
        }
        model = xgb.train(xgb_params, dtrain, num_boost_round=100)
        trained_models[name] = model 
    else:
        model.fit(X_train_scaled, y_train)
        trained_models[name] = model 

# Train meta-model
meta_model = LogisticRegression(random_state=42)
meta_model.fit(meta_features, y_train)


# Make predictions on test set
test_meta_features = np.zeros((x_test.shape[0], len(trained_models)))
for i, (name, model) in enumerate(trained_models.items()):
    if name == 'lgbm':
        test_meta_features[:, i] = model.predict(x_test)
    elif name == 'xgb':
        dtest = xgb.DMatrix(x_test)
        test_meta_features[:, i] = model.predict(dtest)
    else:
        test_meta_features[:, i] = model.predict_proba(X_test_scaled)[:, 1]

final_predictions = meta_model.predict(test_meta_features)
````

````python
import pandas as pd
import numpy as np

u = "18393;2;168;62.0;110;80;1;1;0;0;1"  #Raw Input here

columns = ['age', 'gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']
new_data = [float(x) if '.' in x else int(x) for x in u.split(";")]

data = pd.DataFrame([new_data], columns=columns)
data['age_years'] = (data['age'] / 365).astype(int)

# age categories
data['age_category'] = pd.cut(data['age_years'], 
                            bins=[0, 30, 50, 60, np.inf], 
                            labels=['Under 30', '30-49', '50-59', '60 and above'])

# Encode cholesterol and glucose levels
data['cholesterol_category'] = data['cholesterol'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})
data['glucose_category'] = data['gluc'].map({1: 'Normal', 2: 'Above Normal', 3: 'Well Above Normal'})

# Calculate and categorize BMI
data['bmi'] = data['weight'] / ((data['height'] / 100) ** 2)
data['bmi_category'] = pd.cut(data['bmi'], 
                            bins=[0, 18.5, 25, 30, np.inf], 
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Categorize blood pressure
def bp_category(systolic, diastolic):
    if systolic < 120 and diastolic < 80:
        return 'Normal'
    elif 120 <= systolic < 130 and diastolic < 80:
        return 'Elevated'
    elif 130 <= systolic < 140 or 80 <= diastolic < 90:
        return 'Hypertension Stage 1'
    elif systolic >= 140 or diastolic >= 90:
        return 'Hypertension Stage 2'
    else:
        return 'Hypertensive Crisis'

data['bp_category'] = data.apply(lambda x: bp_category(x['ap_hi'], x['ap_lo']), axis=1)

data['gender'] = data['gender'].map({1: 'Female', 2: 'Male'})
data['smoke'] = data['smoke'].map({0: 'Non-smoker', 1: 'Smoker'})
data['alco'] = data['alco'].map({0: 'Non-drinker', 1: 'Drinker'})
data['active'] = data['active'].map({0: 'Inactive', 1: 'Active'})

columns_to_drop = ['age']
data = data.drop(columns=columns_to_drop)
data
````
